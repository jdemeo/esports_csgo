```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.style as style
style.use('fivethirtyeight')
```

#### Goal: descriptive statistical analysis of income by pr CS:GO Team over specific time periods.

```python
# Get data
money_df = pd.read_csv('data/team_income_190520.csv')
print(money_df.shape)
money_df.head()
```

```python
# Create additional feature
money_df['Avg Winnings per Tourny'] = money_df['TotalUSDPrize'] / money_df['TotalTournaments']
money_df.head()
```

### Look at the top paid teams since 2012

```python
# Top cumulative winnings
top_20 = money_df.sort_values(by=['TotalUSDPrize'], ascending=False).reset_index(drop=True)
top_20 = top_20.loc[:20,]

x_feature = 'TeamName'
y_feature = 'TotalUSDPrize'

plt.bar(top_20[x_feature], top_20[y_feature])
plt.xticks(rotation=90)
plt.xlabel(x_feature)
plt.ylabel(y_feature)
plt.show()
```

```python
sns.set()
# Top Average Winnings
top_20 = money_df.sort_values(by=['Avg Winnings per Tourny'], ascending=False).reset_index(drop=True)
top_20 = top_20.loc[:20,]

x_feature = 'TeamName'
y_feature = 'Avg Winnings per Tourny'

plt.bar(top_20[x_feature], top_20[y_feature])
plt.xticks(rotation=90)
plt.xlabel(x_feature)
plt.ylabel(y_feature)
plt.show()
```

```python
# Same exercise but exclude teams with low tournament count
mini_df = money_df.loc[money_df['TotalTournaments'] > 10,]

# Top Average Winnings
top_20 = mini_df.sort_values(by=['Avg Winnings per Tourny'], ascending=False).reset_index(drop=True)
top_20 = top_20.loc[:20,]

x_feature = 'TeamName'
y_feature = 'Avg Winnings per Tourny'

plt.bar(top_20[x_feature], top_20[y_feature])
plt.xticks(rotation=90)
plt.xlabel(x_feature)
plt.ylabel(y_feature)
plt.show()
```
